```python
import pandas as pd
import numpy as np

# data_xls = pd.read_excel('Demographics.xlsx', 'Demographics', dtype=str, index_col=None)
# data_xls.to_csv('demographics.csv', encoding='utf-8', index=False)
```

```python
#read CSV
demographics = pd.read_csv('Demographics.csv')
demographics
```

```python
#Taking a look at where the NaN values are coming from
demographics.isnull().sum()
```

```python
#dropping NaN values
drop_na = demographics.dropna()
```

```python
dropped_df = drop_na[(drop_na.Gender != 'All Genders') & (drop_na.Injury_Intent != 'All Firearm Deaths') & (drop_na.Race_Ethnicity != 'All Races/Ethnicities')]
dropped_df
```

```python
#Number formatting
# dropped_df.loc[:, "Total_Deaths"] = dropped_df["Total_Deaths"].map('{:.2f}'.format)
pd.options.display.float_format = '{:,.0f}'.format

dropped_df.style.format({'Total_Deaths': '{:,.0f}'.format})
dropped_df
```

```python
#dropping 'First_Year' and 'Last_Year' columns, irrelevant to ML and can be described otherwise
dropped_df.drop(['First_Year', 'Last_Year'], axis = 1, inplace = True)
```

```python
dropped_df.head()
```

```python
x_dummies = pd.get_dummies(dropped_df, columns=['Injury_Intent', 'Race_Ethnicity', 'Gender'])
print(x_dummies.columns)
x_dummies
```

```python
df_setindex = x_dummies.set_index('State')
```

```python
df_setindex.describe()
```

```python
#standardize the data 
from sklearn import preprocessing
standard = preprocessing.scale(df_setindex)
print(standard)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 8))
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(standard)
    wcss.append(kmeans.inertia_) #criterion based on which K-means clustering works
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#k=10 to determine clusters?
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 15, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(standard)

y_kmeans
```

```python
#beginning of  the cluster numbering with 1 instead of 0
y_kmeans1=y_kmeans+1

# New list called cluster
cluster = list(y_kmeans1)
# Adding cluster to our data set
df_setindex['cluster'] = cluster
```

```python
#Mean of clusters 1 to 4
kmeans_mean_cluster = pd.DataFrame(round(df_setindex.groupby('cluster').mean(),1))
kmeans_mean_cluster
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.scatterplot(x=df_setindex['Injury_Intent_Firearm Suicide'], y=df_setindex['Gender_Female'], hue=y_kmeans1)
```
